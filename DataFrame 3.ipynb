## .loc[] method to extract rows by index label

```python
#Let's make 'Film' as our index, and sort the values bu index, so pandas searching become more efficient.
df = pd.read_csv("jamesbond.csv", index_col="Film")
df.sort_index(inplace=True)
df.head(3)
```

```python
#If you pass a single value to .loc[], it returns a Series object for that row. If that index does not exist
#then it raises an error.
df.loc["Dabbang"] #This will raise an error, since there is no movie by this name.
df.loc["Moonraker"]
```

```python
#If some value exist twice in index, then a dataframe is returned
df.loc["Casino Royale"]
```

```python
#Pass a list to extract data for multiple movies
df.loc[["Moonraker", "GoldenEye"]]
```

```python
#Using slice to extract data using index label
df.loc["Casino Royal" : "GoldenEye"] #both indexes are inclusive
```

```python
#If some of the index passed in the list does not exist in the index, then all NaN's are returned for that index.
df.loc[["Casino Royale", "GoldenEye", "Dabbang"]]
```

## .iloc[] method
- even if you set some column as an index, it still gets assigned some index position behind the scenes.
- we can use this index position to extract the data.
- Remember: Unlike .loc[], when we pass a slice (Eg: 2:6), last index position is exclusive in result
- Remember: Unlike .iloc[], if we pass an invalid position in a index position list, it will generate an error.

```python
df.head(3)
```

```python
df.iloc[1]
```

```python
df.iloc[ [2,5,6] ]
```

```python
df.iloc[0:4]
```

```python
df.iloc[[0,2,100]] #Error, since 100 is not a valid index position
```

## .ix[] method
- this is a combination of .loc[] and .iloc[] methods.
- both 'index labels' and 'index postions' can be used to access the data. 

```python
df.ix["GoldenEye"]
```

```python
df.ix[ ["Moonraker", "GoldenEye"] ]
```

```python
df.ix[0:5]
```

```python
df.loc[ ["Casino Royale", "Moonraker", "GoldenEye"], "Actor" : "Budget" ]
```

## Filtering data using .where() method
- It will provide the entire dataframe but only show values in rows which matches the conditions
- All other rows will contain NaNs 

```python
#Normally this is going to return the subset.
mask = df["Actor"] == "Sean Connery"
df[mask]
```

```python
#where will return the entire dataframe, with NaNs in the rows which don't satisfy the conditions
df.where(mask)
```

```python
#use 'other' argument to proide the replacement for the NaN values
df.where(mask, other=False)
```

## Querying data using .query() method.
- Won't work with columns having spaces in its names.
- Works by giving a string in expression.
- You can refer variables using @ sign in expression
- You can use python keywords in expression as well. Eg: and, or, in, not in, lists etc

```python
df = pd.read_csv("jamesbond.csv", index_col="Film")
df.sort_index(inplace=True)
df.head(3)
```

```python
df.columns = [col_name.replace(" ", "_") for col_name in df.columns]
df.columns
```

```python
actor = "Roger Moore"
df.query("Actor == @actor")
```

```python
df.query("Actor == 'David Niven' or Actor == @actor")
```

```python
df.query("Actor in ['David Niven', 'Roger Moore']")
```

## .apply() method on each row of dataframe
- pass axis = 1 or 'columns', which means values are going to get passed along column axis -> ideally meaning rows will be passed.

```python
def good_movie(row):
    actor = row[1]
    budget = row[2]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
```

```python
df.apply(good_movie, axis = 1)
```

## .copy() method
- Use this method when you don't want to refer to original DF or Series.
- This will prevent you from making changes to original Df.

```python
df = pd.read_csv("jamesbond.csv", index_col="Film")
df.sort_index(inplace=True)
df.head(3)
```

```python
#Suppose you extracted a series from DF and made some changes to it. This will also get reflected back in original DF.
actors = df['Actor']
actors.head(3)
```

```python
actors['A View to a Kill'] = 'Sir Roger Moore'
```

```python
#Let's check the original Df as well, changes would have reflected there as well.
df.head(3)
```

```python
#To avoid this, make a copy and then work over it
```

```python
actors = df['Actor'].copy()
```

```python
actors['Casino Royale'] = 'Sir Daniel Craig'
```

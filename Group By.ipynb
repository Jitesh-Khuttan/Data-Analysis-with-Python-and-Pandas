## Some Basic Operations with groupy Object
- size [function]:- gives the total numbers rows in each group of groupby objecty.
- first [function]:- gives the first rows which were encountered while creating groups.
- last [function]:- gives the last rows which were encountered while creating groups.
- groups [attribute]:- returns a dict where grouped column values are keys and list of rows indexes that fall into that group are values.

```python
sectors.size().head() #This is similar to fortunes["Sector"].value_counts()
```

```python
sectors.first().head()
```

```python
sectors.last().head()
```

```python
sectors.groups
```

## Methods on the Groupby Object and DataFrame Columns
- max, min, sum, mean etc.
- applying above methods on whole of the groupby object.
- applying above methods on particular column(s) of the group by object.

```python
#min and max applied on the groupby object will return a DF by extracting min/max values in 
#the left most column
sectors.min().head()  #---> returns min values based on Company column
```

```python
sectors.max().head()
```

```python
#Mathematical functions will only work on numeric columns
sectors.sum().head()   #----> will return the sum for all numbers columns for all groups
sectors.mean().head()
```

```python
#If you want to perform some function on particular column for all groups, then use the following
sectors["Revenue"].min()
sectors["Revenue"].max()
sectors["Employees"].sum()  #------> Total employees working in each group
sectors["Revenue"].mean()   #------> Average number of employees working in each group
```

```python
#Perform actions on multiple columns for each group
sectors[["Revenue", "Employees"]].mean().head()
```

## Grouping by multiple columns
- pass a list of muultiple columns to groupby method to create group on multiple columns.
- this will create a MultiIndex grouping.

```python
fortunes = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortunes.groupby(["Sector", "Industry"])
fortunes.head()
```

```python
#This will be number of entries in each group based on MultiIndex
sectors.size()
```

```python
#Find the sum of Revenue, Profit, Total Employees in each Industry of the Sector
sectors.sum().head(10)
```

```python
#Check just the revenue of each industry which belong to different sectors
sectors["Revenue"].sum().head(10)
```

## The <code> .agg() </code> method
- can be used to apply different mathematical operations on different columns.
- For eg: you want to apply 'sum' to Employees, 'mean' and 'sum' to Revenue, 'max' to Profits.

```python
fortunes = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortunes.groupby("Sector")
fortunes.head()
```

```python
#Apply 'max' for Profits, 'mean' for Revenue, 'sum' to Employees
sectors.agg({
    'Revenue' : 'mean',
    'Profits' : 'max',
    'Employees' : 'sum'
}).head()
```

```python
#Apply multiple methods to all numeric columns
sectors.agg(['sum', 'mean', 'size']).head()
```

```python
sectors.agg({
    'Revenue' : ['sum', 'mean'],
    'Profits' : ['mean', 'max']
}).head()
```

## Iterating through groups
- Problem to solve:- Get the Industry with the highest revenue from each Sector
- Let's explore to get the above result using 3 different methods.

```python
fortunes = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortunes.groupby("Sector")
fortunes.head()
```

```python
df = pd.DataFrame(columns=fortunes.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_in_sector = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_sector)
df.head()
```

```python
#Let's do above using an apply and lambda function
sectors.apply(lambda gp: gp.nlargest(1, "Revenue")).head()
```

```python
#Same result without using nlargest method
sectors.apply(lambda gp: gp[gp['Revenue'] == gp['Revenue'].max()]).head()
```
